```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# data stucture [restaurant_id , 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000], #paradise biryani
    [2, 140000, 140000, 160000, 190000], #beijing bites
    [3, 200000, 230000, 260000, 300000], #pizza hub
    [4, 180000, 210000, 240000, 270000], #burgur point
    [5, 160000, 185000, 205000, 230000], #chai point
])
print("==== zomato sales analysis ====")
print("\n sales data shape",sales_data.shape)
print("\n saple data for 1st 3 restau: \n",sales_data[0:3])# [:3] same output given
print("\n saple data for 1st 3 restau: \n",sales_data[:, :3])# first consider row mins row sari chahiye or column 0:3--> 0,1,2 index
print("\n saple data for 1st 3 restau: \n",sales_data[:, 1:])#row full column and column starting 1(index) to full 
```

```python
# Total sales per year

print(np.sum(sales_data,axis=0))

year_Total=np.sum(sales_data[:, 1:],axis=0)# axis=0 fetch rowwise data and give sum because use np.sum metod --> column vise total -->yearly
print("year wise Total: \n" , year_Total)

row_total=np.sum(sales_data[:, 1:],axis=1)# axis=1 fetch column wise data and give sum because use np.sum metod 
print("year wise Total: \n" , row_total)

# Minimum sales per  restaurant

min_sales=np.min(sales_data[:, 1:],axis=1)
print("restaurant wise minimum: \n",min_sales)

# Maximum sales per year

max_sales_year=np.max(sales_data[:,1:],axis=0)
print("year wise maximum: \n",max_sales_year)

#Average sales per restaurant

Avg_sale=np.mean(sales_data[:, 1:],axis=1)
print("average of  restaurant: \n",Avg_sale)

```

## cumulative sales

```python
cumsum=np.cumsum(sales_data[:, 1:] , axis=1)
print(cumsum)
```

## Create Graph OF restaurant

```python
plt.figure(figsize=(8 , 6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales accross all  restaurant")
plt.xlabel("years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```
